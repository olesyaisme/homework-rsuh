```python
#Задача 6.
#Как известно, у квадратного уравнения () либо нет корней вообще, либо есть один корень, либо два. 
#Пользователь вводит коэффициенты a, b, c: 
#ваша программа должна понять, сколько у этого уравнения корней, и вычислить их, если есть хотя бы один. 
#(Погуглите про дискриминант!)

#Как вычислить квадратный корень, кстати? 
#Подсказка: вычисление корня в математике - это то же самое, что возведение в какую-то степень, значит, нам понадобится оператор **.


a = int(input())
b = int(input())
c = int(input())

d = b**2 - 4*a*c
if d > 0:
    print ((-b + d**0.5 ) / 2*a) and ((-b - d**0.5 ) / 2*a)
elif d == 0:
    print (-b / 2*a)
else:
    print ("Нет корней")
```

```python
#Вы собираетесь на вечеринку и оцениваете её крутость по двум параметрам:

#Количество гостей (guests, целое число)
#Количество пицц (pizzas, целое число)
#Если на одного гостя приходится не менее 0.5 пиццы, вечеринка считается отличной. Если хотя бы 0.25 пиццы — нормальной. 
#Если меньше — плохой.

#Задание: Напишите программу, которая запрашивает у пользователя два числа (гостей и пиццы) и сообщает, насколько хороша вечеринка.

guests = int(input())
pizzas = int(input())
if  pizzas / guests >= 0.5:
    print("Вечеринка отличная")
elif pizzas / guests >= 0.25:
    print("Вечеринка хорошая")
else:
    print ("Вечеринка плохая")
```

```python
#Задача 8.
#В математике есть такая штука, как числа Фибоначчи: это набор чисел, где каждое следующее число - сумма двух предыдущих. 
#Например, нулевое число равно нулю, первое - единице, а второе - 0 + 1 = 1, третье - 1 + 1 = 2 и так далее. 
#Для того, чтобы узнать, чему равно n-е число Фибоначчи, существует формула Бине. 
#Найдите ее и попробуйте написать код в питоне, который будет вычислять n-е число Фибоначчи по этой формуле.


import math

n = int(input())
f = ((1 + math.sqrt(5 ** 0,5)) / 2) ** n
f1 = ((1 - math.sqrt(5 ** 0,5)) / 2) ** n
print ((f - f1) / math.sqrt(5))
```
